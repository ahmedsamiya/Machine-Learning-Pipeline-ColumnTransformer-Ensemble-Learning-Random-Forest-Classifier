```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['time']=encoder.fit_transform(df['time'])
```

```python
df['time'].unique()
```

```python
X=df.drop('time',axis=1)
y=df['time']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import OneHotEncoder 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_col=['total_bill','tip','size']
cat_col=['sex','smoker','day']  
```

```python
num_col  
```

```python
num_pipe=Pipeline(steps=[('scaling',SimpleImputer(strategy='median')),('miss_value',StandardScaler())])
cat_pipe=Pipeline(steps=[('enoding',SimpleImputer(strategy='most_frequent')),('imputation',OneHotEncoder())])
```

```python
preprocessor=ColumnTransformer([('num_pipeline',num_pipe,num_col),('categorial_pipeline',cat_pipe,cat_col)])
```

```python
preprocessor
```

```python
print(type(X_train))
```

```python
X_train=preprocessor.fit_transform(X_train)
X_test=preprocessor.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
models={'support vector Classifier':SVC(),
        'decisiontree':DecisionTreeClassifier(),
        'random_forest':RandomForestClassifier()
}
```

```python

def train_evaluate(X_train,X_test,y_train,y_test,models):
    evaluate={}
    for i in range(len(models)):
         model=list(models.values())[i]
         model.fit(X_train,y_train)
         y_pred=model.predict(X_test)
         model_score=accuracy_score(y_test,y_pred)
         evaluate[list(models.keys())[i]]=model_score
    return evaluate
    
```

```python
train_evaluate(X_train,X_test,y_train,y_test,models)
```
